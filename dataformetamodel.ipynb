Check for duplicates on dataset name and errors

```python
duplicates = dfcombined.duplicated(subset=['dataset_name', 'best_loss', 'train_rmse'], keep=False)
print("Duplicate rows based on 'dataset_name' and 'best_loss':\n", dfcombined[duplicates])
```

```python
#there are duplicates so least time taken will be used in this scenario
```

Create least error without OF df

```python
overfitting_threshold = 0.1

#calculate the absolute difference between train_rmse and best_loss
dfcombined['error_difference'] = (dfcombined['train_rmse'] - dfcombined['best_loss']).abs()

#filter out records with overfitting
filtered_df = dfcombined[dfcombined['error_difference'] <= overfitting_threshold]

print("Before removing OF:",dfcombined.shape)
print("After removing OF:", filtered_df.shape)
```

```python
#group by 'dataset_name' and find the index of the minimum 'best_loss' for each group in the filtered df
idx = filtered_df.groupby('dataset_name')['best_loss'].idxmin()

#use these indices to filter the original df
best_records_df = dfcombined.loc[idx].reset_index(drop=True)
```

```python
best_records_df.shape
```

```python
best_records_df['dataset_name'].value_counts()
```

Filter again using least time_taken

```python
#group by 'dataset_name' and find the index of the minimum 'time_taken' for each group in the filtered df
idx = best_records_df.groupby('dataset_name')['time_taken'].idxmin()

#use these indices to filter the original df
final_df = best_records_df.loc[idx].reset_index(drop=True)
```

```python
final_df.shape
```

```python
final_df['dataset_name'].value_counts()
```

```python
#now each dataset only has one record
```

Drop the unneeded columns

```python
final_df.columns
```

```python
final_df = final_df.drop(columns=['num_clients','best_loss','train_rmse','time_taken','error_difference'])
```

```python
#since the experiments were only ran on synthetic data, I will remove non-numeric values in 'dataset_name' from the metafeatures df
```

```python
#convert 'dataset_name' column to numeric, coerce errors to NaN
metafeatures['dataset_name'] = pd.to_numeric(metafeatures['dataset_name'], errors='coerce')

#drop rows where 'dataset_name' is NaN (non-numeric values)
metafeatures = metafeatures.dropna(subset=['dataset_name'])

#convert to int64
metafeatures['dataset_name'] = metafeatures['dataset_name'].astype('Int64')
```

Merge with inner join on 'dataset_name'

```python
merged_df = pd.merge(final_df, metafeatures, on='dataset_name', how='inner')

merged_df.head(20)
```

```python
#remove dataset_name
merged_df = merged_df.drop(columns=['dataset_name'])
```
